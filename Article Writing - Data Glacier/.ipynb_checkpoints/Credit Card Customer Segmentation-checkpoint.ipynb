```python
### Selcting only the existing customers
credit_card_exist = credit_card[credit_card["Attrition_Flag"] == "Existing Customer"]
credit_card_exist.head(3)
```

```python
credit_card_exist["Card_Category"].value_counts()
```

```python
## Dropping unnecessary Columns
credit_card_exist.drop(labels=["CLIENTNUM","Attrition_Flag"],axis=1,inplace=True)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from kneed import KneeLocator
```

```python
# Numerical Columns
num_col1= credit_card_exist.columns[credit_card_exist.dtypes != "object"]
print("Numerical  Columns:", num_col1,"\n\n")

# Categorical Colums
cat_col1 = credit_card_exist.columns[credit_card_exist.dtypes == "object"]
print("Categorical Columns: ", cat_col1)
```

```python
##scaler = StandardScaler()
#credit_card_exist.loc[:,num_col1]= scaler.fit_transform(credit_card_exist.loc[:,num_col1])
#credit_card_exist[num_col1].head(5)
```

```python
for col in cat_col1:
    labelencoder= LabelEncoder()
    credit_card_exist[col]= labelencoder.fit_transform(credit_card_exist[col])
```

```python
credit_card_exist.head(5)
```

```python
kmeans = KMeans(n_clusters=3,n_init=10,max_iter=300,random_state=10)
kmeans.fit(credit_card_exist)
```

```python
kmeans.labels_
```

```python
## Elbow Method for Selecting K
sse = []
for k in np.arange(1,11,1):
    kmeans = KMeans(n_clusters=k,n_init=10,max_iter=300,random_state=10)
    kmeans.fit(credit_card_exist)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(np.arange(1,11,1),sse)
plt.xticks(np.arange(1,11,1))
plt.xlabel("Number of Cluster")
plt.ylabel("SSE")
plt.show()
```

```python
### USig the Kneed PAckahe to Selec the Cluster
K1 = KneeLocator(np.arange(1,11,1),sse,curve="convex",direction="decreasing")
K1.elbow
```

```python
## Elbow Method for Selecting K
silhouette_score = []

for k in np.arange(1,11,1):
    kmeans = KMeans(n_clusters=k,n_init=10,max_iter=300,random_state=10)
    kmeans.fit(credit_card_exist)
    score= silhouette_score(credit_card_exist,kmeans.labels_)
    silhouette_score.append(score)
```
